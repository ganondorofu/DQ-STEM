```python
# 画像のデータをリストに格納
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
tfr = [1.39, 1.39, 1.41, 1.43, 1.42, 1.46, 1.45, 1.43, 1.42, 1.36, 1.34, 1.35, 1.37, 1.34]

# 線形回帰モデルをインポート
from sklearn.linear_model import LinearRegression

# モデルを作成し、データにフィットさせる
model = LinearRegression()
model.fit([[x] for x in years], tfr)

# 2024年から2030年までの予測値を計算し、表示する
for year in range(2024, 2031):
  pred = model.predict([[year]])[0]
  print(f"{year}年の合計特殊出生率の予測値は{pred:.2f}です。")


# 2024年から2030年までの予測値を計算
predictions = {}
for year in range(2023, 2031):
    pred = model.predict([[year]])[0]
    predictions[year] = pred

import numpy as np

# 予測値とその不確かさを計算
future_years = np.arange(2023, 2031).reshape(-1, 1)
predicted_tfr = model.predict(future_years)

# 不確かさの計算
# 平均二乗誤差の平方根を不確かさの推定値として使用
mse = np.mean((model.predict([[x] for x in years]) - tfr) ** 2)
uncertainty = np.sqrt(mse)

# 上下限の計算
upper_bound = predicted_tfr + uncertainty
lower_bound = predicted_tfr - uncertainty

# 再度プロット
plt.figure(figsize=(10, 6))
plt.plot(years, tfr, color='blue', label='合計特殊出生率')
plt.plot(future_years, predicted_tfr, color='red', label='合計特殊出生率_予測')
plt.fill_between(future_years.flatten(), lower_bound, upper_bound, color='red', alpha=0.2)
plt.title('Total Fertility Rate Prediction with Uncertainty (2010-2030)')
plt.xlabel('Year')
plt.ylabel('Total Fertility Rate')
plt.legend()
```

```python
# 画像のデータをリストに格納
tfr = [1.3, 1.39, 1.41, 1.43, 1.42, 1.46, 1.45, 1.43, 1.42, 1.36] # 合計特殊出生率
gor = [69.8, 69.5, 69.3, 69.1, 68.9] # 外出率

# 外出率のデータが不足しているので、線形補間で埋める
from scipy.interpolate import interp1d
x = [2010, 2013, 2016, 2019, 2022] # 外出率のデータがある年
y = gor # 外出率のデータ
f = interp1d(x, y, kind='linear') # 線形補間関数
x_new = list(range(2010, 2023)) # 2010年から2022年までの年
y_new = f(x_new) # 補間された外出率のデータ

# 合計特殊出生率のデータと同じ範囲にスライス
y_new = y_new[:10] # 2010年から2019年までの外出率のデータ

# 合計特殊出生率と外出率の相関係数を計算
import numpy as np
corr = np.corrcoef(tfr, y_new)[0, 1] # 相関係数
print(f"合計特殊出生率と外出率の相関係数は{corr:.2f}です。")

# 線形回帰モデルで合計特殊出生率を予測
from sklearn.linear_model import LinearRegression
X = np.array(y_new).reshape(-1, 1) # 外出率を説明変数に
y = np.array(tfr) # 合計特殊出生率を目的変数に
model = LinearRegression() # 線形回帰モデル
model.fit(X, y) # モデルの学習
print(f"回帰式はy = {model.coef_[0]:.2f}x + {model.intercept_:.2f}です。")

# 2024年から2030年までの外出率を仮定
gor_future = [68.7, 68.5, 68.3, 68.1, 67.9, 67.7, 67.5] # 外出率の仮定値

# 仮定した外出率に基づいて合計特殊出生率を予測
tfr_future = model.predict(np.array(gor_future).reshape(-1, 1)) # 合計特殊出生率の予測値
print("2024年から2030年までの合計特殊出生率の予測値は以下の通りです。")
for i in range(7):
    print(f"{2024 + i}年: {tfr_future[i]:.2f}")
```
