### **問題**
日本の都道府県別の高齢化率（65歳以上人口の割合）を予測し、2030年に最も高齢化が進む都道府県を特定せよ。また、その結果を視覚化せよ。
ただし、生成AI（gpt-4o、claude等）を用いてコードを作成した場合、使用した生成AIを明記し、どのような工夫をしたかできるだけ詳しく説明してください。

```python
# 必要なライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib  # 日本語フォントの表示
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# データの読み込み
df = pd.read_csv("DQ_人口データセット.csv", encoding="cp932")

# '性別'が'計'のデータのみを使用
df_total = df[df['性別'] == '計'].copy()

# '65歳以上人口'を計算
age_columns_over65 = ['65歳～69歳', '70歳～74歳', '75歳～79歳', '80歳～84歳', '85歳～89歳', '90歳～94歳', '95歳～99歳', '100歳以上']

# 年齢列が数値型でない場合は、数値型に変換（文字列から数値への変換）
for col in age_columns_over65:
    df_total[col] = pd.to_numeric(df_total[col], errors='coerce')

# '65歳以上人口'の列を作成
df_total['65歳以上人口'] = df_total[age_columns_over65].sum(axis=1)

# 都道府県・年ごとの総人口と65歳以上人口を集計
population_data = df_total.groupby(['都道府県名', 'year'])[['総数', '65歳以上人口']].sum().reset_index()

# 高齢化率を計算
population_data['高齢化率'] = population_data['65歳以上人口'] / population_data['総数']

# データの整形
pivot_data = population_data.pivot(index='year', columns='都道府県名', values='高齢化率')

# 年を整数型に変換
pivot_data.index = pivot_data.index.astype(int)

# 年の昇順にソート
pivot_data = pivot_data.sort_index()

# 欠損値の補完
pivot_data = pivot_data.interpolate(method='linear', limit_direction='both', axis=0)

# データの正規化
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(pivot_data)
scaled_df = pd.DataFrame(scaled_data, index=pivot_data.index, columns=pivot_data.columns)

# 時系列データセットの作成
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_steps):
        dataX.append(dataset[i:(i + time_steps)].values)
        dataY.append(dataset.iloc[i + time_steps].values)
    return np.array(dataX), np.array(dataY)

# タイムステップの設定
time_steps = 3

# データセットの作成
X, y = create_dataset(scaled_df, time_steps)

# 年のリストを取得
years = scaled_df.index.tolist()

# 学習データとテストデータの分割（年ベースで80%:20%に分割）
train_size = int(len(years) * 0.8)
train_years = years[:train_size]
test_years = years[train_size:]

# インデックスを取得
train_indices = [years.index(year) for year in train_years if year in years]
test_indices = [years.index(year) for year in test_years if year in years]

# データの分割
X_train = X[train_indices[0]:train_indices[-1]+1]
y_train = y[train_indices[0]:train_indices[-1]+1]
X_test = X[test_indices[0]:test_indices[-1]+1]
y_test = y[test_indices[0]:test_indices[-1]+1]

# LSTMモデルの構築と訓練
model = Sequential()
model.add(LSTM(50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(y_train.shape[1]))
model.compile(loss='mean_squared_error', optimizer='adam')

# モデルの訓練
history = model.fit(X_train, y_train, epochs=50, batch_size=1, validation_data=(X_test, y_test), verbose=2)

# モデルの評価
# 学習曲線のプロット
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='訓練データの損失')
plt.plot(history.history['val_loss'], label='テストデータの損失')
plt.title('モデルの損失の推移')
plt.xlabel('エポック')
plt.ylabel('損失')
plt.legend()
plt.show()

# 将来の高齢化率の予測
last_data = scaled_df.values[-time_steps:]
future_predictions = []

# 予測する年数（2030年まで）
last_year_in_data = scaled_df.index.max()
future_years = list(range(last_year_in_data + 1, 2031))

for year in future_years:
    input_data = last_data.reshape(1, time_steps, -1)
    pred = model.predict(input_data)
    future_predictions.append(pred[0])
    last_data = np.vstack([last_data[1:], pred])

# 予測結果を元のスケールに戻す
future_predictions_inv = scaler.inverse_transform(future_predictions)

# 予測結果をデータフレームに変換
future_df = pd.DataFrame(future_predictions_inv, index=future_years, columns=pivot_data.columns)

# 過去のデータと予測結果を連結
total_df = pd.concat([pivot_data, future_df])

# 2030年に最も高齢化が進む都道府県の特定
aging_rate_2030 = total_df.loc[2030]
max_prefecture = aging_rate_2030.idxmax()
max_rate = aging_rate_2030.max()

print(f"2030年に最も高齢化が進む都道府県は {max_prefecture} で、高齢化率は {max_rate:.2%} です。")

# 結果の視覚化
plt.figure(figsize=(12, 8))
top5_prefectures = aging_rate_2030.sort_values(ascending=False).head(5).index

for prefecture in top5_prefectures:
    plt.plot(total_df.index, total_df[prefecture], label=prefecture)

plt.title('上位5都道府県の高齢化率推移（予測含む）')
plt.xlabel('年')
plt.ylabel('高齢化率')
plt.legend()
plt.grid(True)
plt.show()

# 全都道府県の高齢化率のヒートマップ表示
plt.figure(figsize=(15, 10))
sns.heatmap(total_df.T, cmap='viridis')
plt.title('都道府県別高齢化率の推移（予測含む）')
plt.xlabel('年')
plt.ylabel('都道府県')
plt.show()
    
```
